```python
import pandas as pd
from pathlib import Path

model = "Llama-3.1-8B-Instruct"

input_dir = Path(f"../../data/results/{model}/batches")
output_dir = Path(f"../../data/results/{model}")
```

```python
batch_files = sorted(input_dir.glob("batch_*.pqt"))
dfs = [pd.read_parquet(f) for f in batch_files]
merged_df = pd.concat(dfs, ignore_index=True)
# drop 
merged_df.to_parquet(output_dir / "all_responses_merged.pqt")
```
